**1**. **Insertion** **Sort** :


  **Objective** : To sort a given list of elements in ascending order using the insertion sort technique.

```python
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j = j - 1
        arr[j + 1] = key
    return arr

arr = [12, 11, 13, 5, 6]
print("Original array:", arr)
sorted_arr = insertion_sort(arr)
print("Sorted array:", sorted_arr)

```

2. **Bubble** **Sort**:

     **Objective** : To implement the bubble sort algorithm to sort a given list of elements.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

arr = [12, 11, 5, 7, 2]
print("Original array:", arr)
sorted_arr = bubble_sort(arr)
print("Sorted array:", sorted_arr)
```

3. **Quick** **Sort**:



     **Objective** : To apply the quick sort algorithm for sorting an array by partitioning around a pivot element.


```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        left = [x for x in arr[1:] if x <= pivot]
        right = [x for x in arr[1:] if x > pivot]
        return quick_sort(left) + [pivot] + quick_sort(right)
def start_quick_sort():
    data = [8, 3, 1, 7, 0, 10, 2]
    print("Original array:", data)
    sorted_data = quick_sort(data)
    print("Sorted array:", sorted_data)
start_quick_sort()
```

4. **Merge** **Sort**:


    **Objective** : To implement the merge sort algorithm using the divide-and-conquer approach for efficient sorting.

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

arr = [12, 5, 7, 3, 15, 2]
merge_sort(arr)
print("Sorted array:", arr)
```

5. **Counting** **Sort**:


    **Objective** : To implement the counting sort algorithm to sort elements by counting their occurrences and placing them in the correct sorted position.

```python
def counting_sort(A, k):
    n = len(A)
    C = [0] * (k + 1)
    for j in range(n):
        value = A[j]
        if value < 0 or value > k:
            raise ValueError("Elements must be in range 0..k")
        C[value] = C[value] + 1
    for i in range(1, k + 1):
        C[i] = C[i] + C[i - 1]
    B = [0] * n
    for j in range(n - 1, -1, -1):
        value = A[j]
        B[C[value] - 1] = value
        C[value] = C[value] - 1
    return B
if __name__ == "__main__":
    A = [5, 2, 0, 2, 1, 5]
    k = 5
    print("Original:", A)
    print("Sorted:  ", counting_sort(A, k))
```

8. **Fractional** **Knapsack** :


    **Objective** : To implement the fractional knapsack algorithm using the greedy method to maximize total profit by selecting items based on profit-to-weight ratio.

```python
def fractional_knapsack(weights, values, capacity):
    n = len(values)
    ratio = [(values[i] / weights[i], weights[i], values[i]) for i in range(n)]
    ratio.sort(reverse=True)
    total_value = 0.0
    for r, w, v in ratio:
        if capacity >= w:
            capacity -= w
            total_value += v
        else:
            fraction = capacity / w
            total_value += v * fraction
            break
    return total_value

weights = [10, 20, 30]
values = [60, 100, 120]
capacity = 50
result = fractional_knapsack(weights, values, capacity)
print("Maximum value:", result)
```

9. **Activity** **Selection** *Problem* :


    **Objective** : To implement the activity selection algorithm using a greedy strategy

```python
def activity_selection(start, finish):
    n = len(start)
    activities = list(range(n))
    activities.sort(key=lambda x: finish[x])
    selected = [activities[0]]
    last = activities[0]
    for i in activities[1:]:
        if start[i] >= finish[last]:
            selected.append(i)
            last = i
    return selected

start = [1, 3, 0, 5, 8, 5]
finish = [2, 4, 6, 7, 9, 9]
result = activity_selection(start, finish)
print("Selected activities index:", result)
```

10. **Huffman** **Coding** :


    **Objective** : To construct a Huffman tree and derive variable-length codes that minimize the total encoded message size.

```python
import heapq

class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None
    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(chars, freqs):
    heap = []
    for i in range(len(chars)):
        heapq.heappush(heap, Node(chars[i], freqs[i]))
    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)
        merged = Node(None, left.freq + right.freq)
        merged.left = left
        merged.right = right
        heapq.heappush(heap, merged)
    return heap[0]

def generate_codes(root, code="", codes=None):
    if codes is None:
        codes = {}
    if root.char is not None:
        codes[root.char] = code
    else:
        generate_codes(root.left, code + "0", codes)
        generate_codes(root.right, code + "1", codes)
    return codes

chars = ['A','B','C','D','E','F']
freqs = [5, 9, 12, 13, 16, 45]

root = build_huffman_tree(chars, freqs)
codes = generate_codes(root)
print("Huffman Codes:", codes)
```

11. **Kruskal**'s **Algorithm** :


    **Objective** :To construct a minimum spanning tree by selecting the smallest weight edges while ensuring that no cycles are formed.

```python
class DisjointSet:
    def __init__(self, vertices):
        self.parent = [i for i in range(vertices)]
        self.rank = [0] * vertices

    def find(self, node):
        if self.parent[node] != node:
            self.parent[node] = self.find(self.parent[node])
        return self.parent[node]

    def union(self, u, v):
        rootU = self.find(u)
        rootV = self.find(v)

        if self.rank[rootU] > self.rank[rootV]:
            self.parent[rootV] = rootU
        elif self.rank[rootU] < self.rank[rootV]:
            self.parent[rootU] = rootV
        else:
            self.parent[rootV] = rootU
            self.rank[rootU] += 1
class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []

    def add_edge(self, u, v, weight):
        self.graph.append([u, v, weight])

    def kruskal(self):
        self.graph = sorted(self.graph, key=lambda item: item[2])
        mst = []
        ds = DisjointSet(self.V)

        for edge in self.graph:
            u, v, weight = edge
            if ds.find(u) != ds.find(v):
                mst.append([u, v, weight])
                ds.union(u, v)

        return mst
g = Graph(4)
g.add_edge(0, 1, 10)
g.add_edge(0, 2, 6)
g.add_edge(0, 3, 5)
g.add_edge(1, 3, 15)
g.add_edge(2, 3, 4)

minimum_spanning_tree = g.kruskal()
print("Edges in the Minimum Spanning Tree:")
for edge in minimum_spanning_tree:
    print(edge[0], "-", edge[1], ":", edge[2])
```

12. **Prim**'**s** **Algorithm** :


    **Objective** : To implement Prim’s algorithm to find the Minimum Spanning Tree (MST) of a connected weighted graph by selecting edges with the minimum weight that connect new vertices to the growing tree.

```python
import sys

def prims(graph):
    n = len(graph)
    selected = [False] * n
    selected[0] = True
    mst = []
    total_cost = 0

    for _ in range(n - 1):
        minimum = sys.maxsize
        x = y = 0
        for i in range(n):
            if selected[i]:
                for j in range(n):
                    if not selected[j] and graph[i][j] != 0 and graph[i][j] < minimum:
                        minimum = graph[i][j]
                        x = i
                        y = j
        mst.append((x, y, graph[x][y]))
        total_cost += graph[x][y]
        selected[y] = True

    return mst, total_cost

graph = [
    [0, 2, 0, 6, 0],
    [2, 0, 3, 8, 5],
    [0, 3, 0, 0, 7],
    [6, 8, 0, 0, 9],
    [0, 5, 7, 9, 0]
]

mst, cost = prims(graph)
print("MST:", mst)
print("Total cost:", cost)
```

13. **Dijkstra’s** **Algorithm** :


    **Objective** : To compute the minimum distance from a source node to every other node in a graph using Dijkstra’s greedy approach.

```python
import heapq

def dijkstra(graph, start):
    n = len(graph)
    dist = [float('inf')] * n
    dist[start] = 0
    pq = [(0, start)]
    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v, w in graph[u]:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))
    return dist

graph = {
    0: [(1, 4), (2, 1)],
    1: [(3, 1)],
    2: [(1, 2), (3, 5)],
    3: []
}

distances = dijkstra(graph, 0)
print("Shortest distances from source:", distances)
```

14. **Bellman–Ford** **Algorithm** :


    **Objective** : To find shortest paths in a graph with positive or negative edge weights using the relaxation technique of the Bellman–Ford algorithm.

```python
def bellman_ford(n, edges, source):
    dist = [float('inf')] * n
    dist[source] = 0
    for _ in range(n - 1):
        for u, v, w in edges:
            if dist[u] != float('inf') and dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
    for u, v, w in edges:
        if dist[u] != float('inf') and dist[u] + w < dist[v]:
            return None
    return dist

n = 5
edges = [
    (0, 1, -1),
    (0, 2, 4),
    (1, 2, 3),
    (1, 3, 2),
    (1, 4, 2),
    (3, 2, 5),
    (3, 1, 1),
    (4, 3, -3)
]

distances = bellman_ford(n, edges, 0)
print("Shortest distances from source:", distances)
```

15. **0** **1** **KNAPSACK** :

    **Objective** : To understand and implement the 0/1 knapsack solution using a DP table that evaluates optimal choices based on weight and value constraints.

```python
def knapsack(wt, val, W, n):
    dp = [[0 for _ in range(W + 1)] for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(1, W + 1):
            if wt[i - 1] <= w:
                dp[i][w] = max(val[i - 1] + dp[i - 1][w - wt[i - 1]], dp[i - 1][w])
            else:
                dp[i][w] = dp[i - 1][w]
    return dp[n][W]

val = [60, 100, 120]
wt = [10, 20, 30]
W = 50
n = len(val)

result = knapsack(wt, val, W, n)
print("Maximum value:", result)
```

16. **Longest** **Common** **Subsequence** (**LCS**) :

    **Objective** : To implement the Longest Common Subsequence (LCS) algorithm using dynamic programming to find the longest subsequence common to two given strings.

```python
def lcs(X, Y):
    m = len(X)
    n = len(Y)
    dp = [[0 for _ in range(n + 1)] for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if X[i - 1] == Y[j - 1]:
                dp[i][j] = 1 + dp[i - 1][j - 1]
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[m][n]

X = "AGGTAB"
Y = "GXTXAYB"
result = lcs(X, Y)
print("Length of LCS:", result)
```

17. **Matrix** **Chain** **Multiplication** :

    **Objective** : To implement the Matrix Chain Multiplication algorithm using dynamic programming to determine the optimal way of parenthesizing matrix products with minimum computational cost.

```python
import sys
def matrix_chain_multiplication(p):
    n = len(p) - 1
    m = [[0 for _ in range(n + 1)] for _ in range(n + 1)]
    s = [[0 for _ in range(n + 1)] for _ in range(n + 1)]
    for i in range(1, n + 1):
        m[i][i] = 0
    for L in range(2, n + 1):
        for i in range(1, n - L + 2):
            j = i + L - 1
            m[i][j] = sys.maxsize
            for k in range(i, j):
                cost = m[i][k] + m[k+1][j] + p[i-1] * p[k] * p[j]
                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k

    return m[1][n], s
def print_optimal_parens(s, i, j):
    if i == j:
        print(f"A[{i}]", end="")
        return
    print("(", end="")
    print_optimal_parens(s, i, s[i][j])
    print_optimal_parens(s, s[i][j] + 1, j)
    print(")", end="")
if __name__ == "__main__":
    p = [10, 20, 30, 40, 50]

    min_multiplications, s_matrix = matrix_chain_multiplication(p)

    print(f"The minimum number of scalar multiplications needed is: {min_multiplications}")
    print("Optimal Parenthesization is: ", end="")
    print_optimal_parens(s_matrix, 1, len(p) - 1)
    print()
```
