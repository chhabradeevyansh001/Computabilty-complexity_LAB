```python
# BFS
graph = {'A' : ['B','C'],'B' : ['D', 'E'],'C' : ['F'],
  'D' : [],'E' : ['F'],'F' : []}
visited = []
queue = []
def bfs(visited, graph, node):
  visited.append(node)
  queue.append(node)
  while queue:
    s = queue.pop(0)
    print (s, end = " ")
    for neighbour in graph[s]:
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)
bfs(visited, graph, 'A')
```

```python
# DFS
graph = {'5' : ['3','7'],'3' : ['2', '4'],'7' : ['8'],
  '2' : [],'4' : ['8'],'8' : []}
visited = set()
def dfs(visited, graph, node):
    if node not in visited:
        print (node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)
print("Following is the Depth-First Search")
dfs(visited, graph, '5')
```

```python
# Prim's Algorithm - Easy version for understanding
# Graph as adjacency matrix

graph = [
    [0, 1, 5, 0],  # A
    [1, 0, 2, 4],  # B
    [5, 2, 0, 3],  # C
    [0, 4, 3, 0]   # D
]

n = len(graph)
selected = [False] * n   # To mark visited nodes
edge_weight = [999999] * n   # Track cheapest edge to connect each node
parent = [-1] * n        # To remember from which node we connect
edge_weight[0] = 0       # Start from node 0 (A)

print("Starting from node A\n")

for _ in range(n):
    # Step 1: pick node with smallest edge_weight not in MST
    min_cost = 999999
    u = -1
    for i in range(n):
        if not selected[i] and edge_weight[i] < min_cost:
            min_cost = edge_weight[i]
            u = i

    selected[u] = True  # include this node in MST
    print(f"Added node {u} to MST")

    # Step 2: update neighboring nodes
    for v in range(n):
        if graph[u][v] != 0 and not selected[v] and graph[u][v] < edge_weight[v]:
            edge_weight[v] = graph[u][v]
            parent[v] = u
            print(f"  Node {v} can connect through {u} with cost {graph[u][v]}")

print("\nMST edges:")
total = 0
for i in range(1, n):
    print(f"{parent[i]} -> {i}  cost = {graph[i][parent[i]]}")
    total += graph[i][parent[i]]
print("Total cost of MST:", total)

```
